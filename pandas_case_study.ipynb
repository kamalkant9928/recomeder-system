Using the given transactions dataset, calculate the percentage contribution of repeat customers (customers with more than one transaction) to the total revenue, but only for those repeat customers whose average transaction amount is above the overall dataset’s average transaction amount.

```python
x=df.groupby('Customer ID').agg(count=('Transaction ID','count'),total_revenue=('revenue','sum'),avg_transaction=('Transaction Amount','mean')).reset_index()
x['%']=(x['total_revenue'])*100/df['revenue'].sum()
x[(x['count']>1 )&(x['avg_transaction']>df['Transaction Amount'].mean())]['%'].sum()
```

Using the same transactions dataset, find the percentage of total revenue generated by customers whose first transaction amount is higher than the median transaction amount of the entire dataset.

```python
x=df.groupby('Customer ID').agg(total_revenue=('revenue','sum'),first=('Transaction Amount','first')).reset_index()
x['%']=(x['total_revenue']*100)/df['revenue'].sum()
x[x['first']>df['Transaction Amount'].median()]['%'].sum()
```

Using the same transactions dataset, calculate the percentage of total revenue generated by customers whose maximum transaction amount is greater than the 75th percentile of all transaction amounts, but only include customers who have made at least 3 transactions.

```python
x=df.groupby('Customer ID').agg(count=('Transaction Amount','count'),total_revenue=('revenue','sum'),max=('Transaction Amount','max')).reset_index()
x['%']=(x['total_revenue']*100)/df['revenue'].sum()
x[(x['count']>=3) &(x['max']>df['Transaction Amount'].quantile(0.75))]['%'].sum()
```

Using the same transactions dataset, calculate the percentage of total revenue contributed by customers whose last transaction amount is lower than their own average transaction amount, and who have made transactions on at least 2 different days.


```python
x=df.groupby('Customer ID').agg(count_days=('Transaction Date','nunique'),total_revenue=('revenue','sum'),last=('Transaction Amount','last'),avg=('Transaction Amount','mean')).reset_index()
x['%']=round((x['total_revenue']*100)/df['revenue'].sum(),2)
x[(x['last']<x['avg']) &(x['count_days']>=2)]['%'].sum()
```

Using the same transactions dataset, calculate the percentage of total revenue contributed by customers whose transaction amounts show strictly increasing behavior over time, and who have made at least 3 transactions.

```python
x=df.groupby('Customer ID').agg(count1=('Transaction ID','count'),total_revenue=('revenue','sum'),diff=('Transaction Amount',lambda s:s.is_monotonic_increasing)).reset_index()
x['%']=(x['total_revenue']*100)/df['revenue'].sum()
x[(x['count1']>=3)&(x['diff']==True)]['%'].sum()
```

Find the percentage of total revenue generated by customers whose

maximum single-transaction revenue is more than 2× their average transaction revenue

AND who have at least 3 transactions

```python
x=df.groupby('Customer ID').agg(max=('revenue','max'),total_revenue=('revenue','sum'),count=('Transaction ID','count'),avg=('revenue','mean')).reset_index()
x['%']=x['total_revenue']*100/df['revenue'].sum()
x[(x['count']>=3)&(x['max']>x['avg']*2)]['%'].sum()
```

Identify customers who show a consistent acceleration in spending over time.

A customer qualifies if:

They have made transactions on at least 4 different dates.

When transactions are ordered by Transaction Date, the difference between consecutive transaction amounts is strictly increasing (i.e., second increase > first increase > third increase … pattern holds).

For these customers, calculate the percentage contribution of their total revenue to the overall dataset revenue.

```python
df=df.sort_values(['Transaction Date','Transaction ID'])
x=df.groupby('Customer ID').agg(count=('Transaction Date','nunique'),total_revenue=('revenue','sum'),incre=('Transaction Amount',lambda s:(s.diff().dropna()>0).all())).reset_index()
x['%']=x['total_revenue']*100/df['revenue'].sum()
x[(x['count']>=4)&(x['incre']==True)]['%'].sum()
```
